```python
import pandas as pd
import re
```

```python
df = pd.read_clipboard(header=None, sep=':')
```

```python
df.columns=['game', 'colors_raw']
```

```python
df
```

```python
# expand into columns
df.colors.str.strip().apply(lambda text: re.split(';|,', text)).apply(lambda x: pd.Series(x))
```

```python
df_expanded = pd.concat([df['game'],
          df.colors_raw.str.strip().apply(lambda text: re.split(';|,', text)).apply(lambda x: pd.Series(x))], axis=1)
```

```python
df_long = df_expanded.melt(id_vars='game').drop(columns='variable').dropna()
```

```python
# extract into separate columns
df_long['color'] = df_long.value.str.split().str[1]
df_long['count'] = df_long.value.str.split().str[0].astype(int)
```

```python
df_long.game = df_long.game.str.replace('Game', '')
```

```python
df_long
```

### Determine which games would have been possible if the bag had been loaded with only 12 red cubes, 13 green cubes, and 14 blue cubes

```python
# identify groups that are outside bounds
df_long_filtered = df_long.groupby(['game', 'color']).filter(lambda x: ((max(x['count']) > 12) & (x['color'] == 'red')).any() |
                                                    ((max(x['count']) > 13) & (x['color'] == 'green')).any() |
                                                    ((max(x['count']) > 14) & (x['color'] == 'blue').any()))
```

```python
df_long_filtered.sort_values(['game', 'color']).head(20)
```

```python
sum([int(x) for x in df_long.game.unique() if x not in df_long_filtered.game.unique()])
```
